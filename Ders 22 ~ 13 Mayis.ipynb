```python
def my_rot(im_1):
    m=im_1.shape[0]
    n=im_1.shape[1]

    my_new_image=np.zeros((n,m,3),dtype=int)
    for row in range(im_1.shape[0]):
        for column in range(im_1.shape[1]):
            my_new_image[column,row,:] = im_1[row,column,:]
#     plt.imshow(my_new_image)
#     plt.show()
    return my_new_image
```

```python
my_rot(im_1)
```

```python

def my_norm(my_v):
    return int(((my_v[0]**2)+(my_v[1]**2)+(my_v[2]**2))**0.5)

def my_norm_1(my_v):
    
    s= int(((my_v[0]**2)+(my_v[1]**2)+(my_v[2]**2))**0.5)
    if s>130:
        return 1
    else:
        return 0
```

```python
def my_convert_rgb_to_gray(im_1):
    m=im_1.shape[0]
    n=im_1.shape[1]

#     my_new_image=np.zeros((n,m,3),dtype=int)
    my_new_image=np.zeros((m,n,3),dtype=int)
    
    for row in range(im_1.shape[0]):
        for column in range(im_1.shape[1]):
            my_new_image[row,column] = my_norm(im_1[row,column,:])
    return my_new_image

def my_convert_rgb_to_bw(im_1):
    m=im_1.shape[0]
    n=im_1.shape[1]

    #my_new_image=np.zeros((n,m,3),dtype=int)
    my_new_image=np.zeros((m,n,3),dtype=int)
    
    for row in range(im_1.shape[0]):
        for column in range(im_1.shape[1]):
            my_new_image[row,column] = my_norm_1(im_1[row,column,:])
    return my_new_image
```

```python
file_name = "tukan.jpeg"
im_1 = plt.imread(file_name)
im_2 = my_rot(im_1)
plt.subplot(1,2,1)
plt.imshow(im_1)
plt.subplot(1,2,2)
plt.imshow(im_2)
```
